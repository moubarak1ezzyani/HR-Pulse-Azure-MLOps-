### Data Cleaning - Salary Estimate

```python
# --- protoype : $137K-$171K (Glassdoor est.)
# remove : (- - -)
salary=df_init['Salary Estimate'].apply(lambda x : x.split('(')[0].strip())

# remove $, K
salary = salary.str.replace('$', '', regex=False).str.replace('K', '', regex=False)

# optional but common
# salary = salary.str.replace('Employer Provided Salary : ', '', regex=False).str.replace('Per Hour', '', regex=False).str.strip()
print (salary) #  OUTPUT  : 137-171

def calculate_avg_salary(sal_string):   # 137-171 => 154
    try:
        if '-' in sal_string:
            min_sal=float(sal_string.split('-')[0])
            max_sal=float(sal_string.split('-')[1])
            avg_sal = (min_sal+max_sal)/2
            return avg_sal
        else:
            return float(sal_string)
        
    except ValueError:
        return None

# new column : Average salary
df_init['Avg_Sal_K'] = salary.apply(calculate_avg_salary) 
print(df_init[['Salary Estimate','Avg_Sal_K']])
```

### Salary Estimate vs Ratings

```python
plt.figure(figsize=(15,5))
valid_sal_rat=df_init[(df_init['Rating']>0) & (df_init['Avg_Sal_K'].notnull())]

sns.scatterplot(data=valid_sal_rat, y='Rating', x='Avg_Sal_K', alpha=0.6, color='purple')
plt.title('Does Higher Salary Mean Better Company Ratings?')
plt.xlabel('Average Salary ($K)')
plt.ylabel('Company Rating (Out of 5.0)')
plt.grid(True, linestyle='--', alpha=0.5)
```

# The NLP Preprocessing
## data loading

```python
nltk.download('punkt', quiet=True)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

## Lowercasing

```python
# text test
text = """
The Data Scientists are DEVELOPING 3 new models! 
They love Python & SQL. It's better than manual analysis.
"""
text = text.lower()
text=re.sub(r'[^a-z\s]', '',text)
```

## Tokenization

```python
tokens=word_tokenize(text)
print(tokens)
```

## Stop Words Removal

```python
stop_words = set(stopwords.words('english'))
tokens = [w for w in tokens if w not in stop_words]

print(f"stop : {stop_words}")
print(f"tokens {tokens}")
```

## Lemmatization (Root finding)

```python
lemmatizer = WordNetLemmatizer()
tokens = [lemmatizer.lemmatize(w) for w in tokens]
print(f"tolen : {tokens}")
```
